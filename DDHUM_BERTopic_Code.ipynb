```python
import pandas as pd
from umap import UMAP
from bertopic import BERTopic
from google.colab import drive
from sentence_transformers import SentenceTransformer
```

# **BERTOPIC for Ads that mention Lula's name**



```python
path_apoio_l13 = "PATH/TO/CSV/FILE"

df_apoio_l13 = pd.read_csv(path_apoio_l13)

df_apoio_l13['BODY_ADS'].apply(lambda x: [item for item in x if item not in stopwords])

sentences_l13 = df_apoio_l13['BODY_ADS'].str.lower().to_list()
dates_l13 = df_apoio_l13['DATA_CRIACAO'].to_list()
```

```python
umap_model = UMAP(random_state=0)

topic_model_l13_TL = BERTopic(language="multilingual", nr_topics="auto", umap_model=umap_model)
```

```python
# Prepare embeddings
sentence_model = SentenceTransformer("distiluse-base-multilingual-cased")
embeddings = sentence_model.encode(sentences_l13, show_progress_bar=False)
```

```python
topic_l13, probs_l13 = topic_model_l13_TL.fit_transform(sentences_l13)
```

```python
topic_model_l13_TL.get_topic_info()
```

```python
topic_model_l13_TL.reduce_topics(sentences_l13, nr_topics=11)
```

```python
topic_model_l13_TL.get_topic_info()
```

```python
#after reduction
topic_model_l13_TL.visualize_barchart(top_n_topics = 10)
```

```python
topics_l13_overtime = topic_model_l13_TL.topics_over_time(sentences_l13, dates_l13, nr_bins=20)
```

```python
#after reduction
topic_model_l13_TL.visualize_topics_over_time(topics_l13_overtime, top_n_topics=10)
```
